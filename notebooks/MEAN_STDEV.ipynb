```python
import pandas as pd
import statistics
import numpy as np
import glob
import os
import git
```

```python
# get all files from Folder Iteration ending with 'json.gz'
event_log_path = '/logs/finished/*.csv'
git_path = git.Repo(".", search_parent_directories=True).git.rev_parse("--show-toplevel")
final_path = git_path+event_log_path
files = glob.glob(final_path)

output_df = pd.DataFrame(columns= ['method','mean','stdev','ds_year','cluster','epochs'])

for filepath in files:

    # load eventlog 
    # event log configuration
    
    event_log_path = filepath
    file_name = os.path.basename(filepath)
    df = pd.read_csv(filepath, sep=';')

    ds_year = file_name.split("_")[1]
    cluster = file_name.split("_")[2]
    if cluster == "k" : cluster = "k_means"
    epochs = file_name.split("_")[-2]

    d = {'method': ['Autoencoder', 'TRACE2VEC', 'CASE2VEC E', 'CASE2VEC E+C', 'GRU', 'LSTM'],
     
     'mean': [df.loc[df['Method'] == 'Autoencoder']["b3"].mean(), 
              df.loc[df['Method'] == 'TRACE2VEC']["b3"].mean(),
              df.loc[df['Method'] == 'CASE2VEC E']["b3"].mean(),
              df.loc[df['Method'] == 'CASE2VEC E+C']["b3"].mean(),
              df.loc[df['Method'] == 'GRU']["b3"].mean(),
              df.loc[df['Method'] == 'LSTM']["b3"].mean()], 
     
     'stdev': [statistics.stdev(df.loc[df['Method'] == 'Autoencoder']["b3"]), 
              statistics.stdev(df.loc[df['Method'] == 'TRACE2VEC']["b3"]),
              statistics.stdev(df.loc[df['Method'] == 'CASE2VEC E']["b3"]),
              statistics.stdev(df.loc[df['Method'] == 'CASE2VEC E+C']["b3"]),
              statistics.stdev(df.loc[df['Method'] == 'GRU']["b3"]),
              statistics.stdev(df.loc[df['Method'] == 'LSTM']["b3"])] }
    df_new = pd.DataFrame(data=d)
    df_new[['ds_year','cluster','epochs']] = [ds_year,cluster,epochs]

    output_df = output_df.append(df_new)


output_df['presentation'] = output_df['mean'].astype(str).str[:5]+" ± "+ output_df['stdev'].astype(str).str[:5]
output_df
output_df.to_csv("Tab2")
```
